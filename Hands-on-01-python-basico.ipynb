### Criando uma função simples (decidir se vale a pena mostrar)

```python
def conta(x,y,z):
    w = x+y+z
    raiz= round(w ** 0.5,2)
    s='a soma dos elementos da ' + str(w)
    u='a raiz da soma dos elementos da ' + str(raiz)
    
    if w >= 100:
        v= 'e essa soma da mais de 100'
    else:
        v= 'e essa soma nao da mais de 100'
        
    print(s,v)
    print(u)
```

```python
conta(a,b,c)
```

```python
conta(a,b,d)
```

### Apply

```python
def check_valor(s):
    if s['valor'] == 50:
        texto = 'ticket baixo'
    
    elif (s['valor'] >50) & (s['valor'] <100):
        texto = 'ticket medio'
    
    else:
        texto = 'ticket alto'
    
    return texto
```

```python
base_vendas6['label_valor'] = base_vendas6.apply(check_valor,axis=1)
```

```python
base_vendas6
```

```python
def test(x):
    if x =='1':
        t="Digito 1"
    else:
        t='Digito nao é 1'
    
    return t        
```

```python
base_vendas6['teste'] = base_vendas6.ult_dig.apply(test)
```
